```python
class Memory:
  def __init__(self):
    self.storage=[]
  def add(self,text):
    self.storage.append(text)
  def get_all(self):
    return self.storage
```

```python
m = Memory()
m.add("Hello")
m.add("World")
print(m.get_all())
```
